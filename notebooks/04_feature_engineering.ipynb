```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv(
    "../outputs/clean_dataset.csv"
)

df.head()
```

```python
features = [
    "latitude",
    "longitude",
    "magnitude",
    "depth"
]
X = df[features]
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

import joblib

joblib.dump(
    scaler,
    "../models/scaler.pkl"
)
```

```python
print(X_scaled.shape)
```

```python
mm = MinMaxScaler()

df["mag_norm"] = mm.fit_transform(
    df[["magnitude"]]
)

df["depth_norm"] = mm.fit_transform(
    df[["depth"]]
)

df["risk_score"] = (
      0.60 * df["mag_norm"]
    + 0.40 * (1 - df["depth_norm"])
)

df["risk_level"] = pd.qcut(
    df["risk_score"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)

print(
    df["risk_level"]
      .value_counts()
)

df.to_csv(
    "../outputs/feature_engineered.csv",
    index=False
)
```
