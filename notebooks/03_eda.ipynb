```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    "../outputs/clean_dataset.csv"
)

df.head()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["magnitude"],
    bins=40,
    kde=True
)

plt.title("Magnitude Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["depth"],
    bins=50,
    kde=True
)

plt.title("Depth Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.scatter(
    df["longitude"],
    df["latitude"],
    s=2,
    alpha=0.4
)

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.title("Earthquake Distribution")
plt.show()
```

```python
yearly = (
    df.groupby("year")
      .size()
      .reset_index(name="count")
)

plt.figure(figsize=(12,5))

sns.lineplot(
    data=yearly,
    x="year",
    y="count"
)

plt.show()
```

```python
numeric_cols = [
    "magnitude",
    "depth",
    "latitude",
    "longitude",
    "phasecount",
    "azimuth_gap"
]

corr = df[numeric_cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=df["magnitude"]
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=df["depth"]
)

plt.show()
```
