```python
import pandas as pd
import numpy as np
df = pd.read_csv("../data/indonesia_earthquake.csv")
print("Shape Awal:")
print(df.shape)
print(df.isnull().sum())

missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

missing_percent = (
    missing / len(df)
) * 100

pd.DataFrame({
    "missing": missing,
    "percent": missing_percent
})
df = df.dropna()
print(df.shape)

print(df.isnull().sum())


before = len(df)

df = df.drop_duplicates()

after = len(df)

print("Sebelum :", before)
print("Sesudah :", after)
print("Terhapus :", before-after)

df["date"].head()
df["date"] = pd.to_datetime(df["date"])
df["time"].head(20)
df["time"].sample(20)
print(df["time"].dtype)
df[["date","time"]].head(20)

df["date"] = pd.to_datetime(df["date"])

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["quarter"] = df["date"].dt.quarter
df.to_csv(
    "../outputs/clean_dataset.csv",
    index=False
)

print(df["magnitude"].describe())

print(df["depth"].describe())
print(df.shape)
```
