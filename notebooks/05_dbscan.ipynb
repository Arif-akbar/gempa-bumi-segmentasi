```python
import pandas as pd

df = pd.read_csv(
    "../outputs/feature_engineered.csv"
)
```

```python
import joblib

scaler = joblib.load(
    "../models/scaler.pkl"
)
```

```python
features = [
    "latitude",
    "longitude",
    "magnitude",
    "depth"
]

X = df[features]

X_scaled = scaler.transform(X)

from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

neighbors = NearestNeighbors(
    n_neighbors=10
)

neighbors_fit = neighbors.fit(
    X_scaled
)

distances, indices = (
    neighbors_fit.kneighbors(
        X_scaled
    )
)

distances = np.sort(
    distances[:,9]
)

plt.figure(figsize=(10,5))

plt.plot(distances)

plt.title(
    "K Distance Graph"
)

plt.show()
```
